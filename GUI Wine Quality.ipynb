```python
from tkinter import *
import pickle
from tkinter import ttk
import numpy as np
import warnings
warnings.filterwarnings('ignore')



root=Tk()
root.title('Wine Quality')
root.geometry('1350x800')
Label(root,text='Wine Quality Prediction',bg='red',height=1,width=200,
      font=('Times',20,'bold')).pack(side='top')

def predict():
    f=open('scaling','rb')
    sc=pickle.load(f)
    f.close()
    f1=open('Red','rb')
    clf1=pickle.load(f1)
    f1.close()
    f2=open('white','rb')
    clf2=pickle.load(f2)
    f2.close()
    winetype=(winetype_var.get()).lower()
    a1=a1_var.get()
    a2=a2_var.get()
    a3=a3_var.get()
    a4=a4_var.get()
    a5=a5_var.get()
    a6=a6_var.get()
    a7=a7_var.get()
    a8=a8_var.get()
    a9=a9_var.get()
    a10=a10_var.get()
    a11=a11_var.get()
    winetype_var.set('')
    a1_var.set('')
    a2_var.set('')
    a3_var.set('')
    a4_var.set('')
    a5_var.set('')
    a6_var.set('')
    a7_var.set('')
    a8_var.set('')
    a9_var.set('')
    a10_var.set('')
    a11_var.set('')

    new=np.array([[a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11]])
    a=sc.transform(new)
    if winetype=='red':
        pred=clf1.predict(a)
    elif winetype=='white':
        pred=clf2.predict(a)
    else:
        l1.config(text='Please Select Wine Type')
        
    pred=pred[0]
    if pred>0.5:
        l1.config(text='Good Quality Wine')
    else:
        l1.config(text='Bad Quality Wine')    
   
            


frame=LabelFrame(root,bd=5,text='Characterstics',
                     pady=10,padx=10,font=('times',12,'bold'))
frame.place(x=15,y=60,height=400,width=1250)

winetype=Label(frame,text='Wine Type')
winetype.grid(row=1,column=6)
winetype_var=StringVar()
winetypecombo=ttk.Combobox(frame,font=('times',10),width=17,textvariable=winetype_var)
winetypecombo['values']=('Red','White')
winetypecombo.grid(row=1,column=7)


a1=Label(frame,text='Fixed Acidity').grid(row=1,column=2)
a1_var=DoubleVar()
Entry(frame,textvariable=a1_var).grid(row=1,column=4)

a2=Label(frame,text='Volatile Acidity').grid(row=2,column=2)
a2_var=DoubleVar()
Entry(frame,textvariable=a2_var).grid(row=2,column=4)

a3=Label(frame,text='Citric Acid').grid(row=3,column=2)
a3_var=DoubleVar()
Entry(frame,textvariable=a3_var).grid(row=3,column=4)

a4=Label(frame,text='Residual Sugar').grid(row=4,column=2)
a4_var=DoubleVar()
Entry(frame,textvariable=a4_var).grid(row=4,column=4)

a5=Label(frame,text='Chlorides').grid(row=5,column=2)
a5_var=DoubleVar()
Entry(frame,textvariable=a5_var).grid(row=5,column=4)

a6=Label(frame,text='Free Sulfur Dioxide').grid(row=6,column=2)
a6_var=DoubleVar()
Entry(frame,textvariable=a6_var).grid(row=6,column=4)

a7=Label(frame,text='Total Sulfur Dioxide').grid(row=7,column=2)
a7_var=DoubleVar()
Entry(frame,textvariable=a7_var).grid(row=7,column=4)

a8=Label(frame,text='Density').grid(row=8,column=2)
a8_var=DoubleVar()
Entry(frame,textvariable=a8_var).grid(row=8,column=4)

a9=Label(frame,text='pH').grid(row=9,column=2)
a9_var=DoubleVar()
Entry(frame,textvariable=a9_var).grid(row=9,column=4)

a10=Label(frame,text='Sulphates').grid(row=10,column=2)
a10_var=DoubleVar()
Entry(frame,textvariable=a10_var).grid(row=10,column=4)

a11=Label(frame,text='Alcohol').grid(row=11,column=2)
a11_var=DoubleVar()
Entry(frame,textvariable=a11_var).grid(row=11,column=4)




Predict=Button(frame,text='Predict',bg='grey',width=20,height=2,command=predict)
Predict.grid(row=13,column=4)


l1=Label(frame,text='Result is here',height=5,width=30,font='bold',bg='skyblue')
l1.place(x=500,y=50)









root.mainloop()
```
